```python
def get_english_urls(translations):
  all_english_urls = [(l['name'],l['download_url']) for l in requests.get(translations).json() if '/en' in l['url']]
  return all_english_urls
```

```python
def get_english_translations(english_urls):# return dict of {'translation_name':'full_translation}
  return {name:requests.get(url).text for name,url in english_urls}
```
